```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("istherecorrelation.csv", delimiter=";", decimal=',')
df
```

```python


correl = df['WO [x1000]'].corr(df['NL Beer consumption [x1000 hectoliter]'])
plt.figure(figsize=(12,8))
plt.plot(df['Year'], df["WO [x1000]"], label = "WO (x1000)", color = 'red')
plt.plot(df['Year'], df['NL Beer consumption [x1000 hectoliter]'],
         label = "NL Beer consumption (x1000 hectoliter)", color ='royalblue')
plt.axhline((df['WO [x1000]']).mean(), linestyle = '--',
            alpha = 0.7, color = 'black', label = 'Mean WO')
plt.axhline((df['NL Beer consumption [x1000 hectoliter]']).mean(),
            linestyle = '--', alpha = 0.7, color= 'green', label = 'Mean Beer consumption')
plt.title(f"The correlation is {round(correl,2)}.")
plt.legend()
plt.savefig(fname = 'my_plot.png', dpi = 300)
```
